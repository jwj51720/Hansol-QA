```python
import pandas as pd
import re
pd.options.display.max_colwidth = 2000
```

```python
test = pd.read_csv("../data/test.csv")
new_test = test.copy()
```

```python
def split_questions(text):
  text = re.sub(r'[\'"]', '', text)
  if "또한," in text:
    return text.split("또한,")
  elif "그리고" in text:
    return text.split("그리고")
  else:
    return [text]

new_test["질문"] = test["질문"].apply(split_questions)
new_test = new_test.explode("질문")
new_test = new_test.reset_index(drop=True)
idx = 0
```

```python
def update_questions(dataframe, id):
    id_value = f"TEST_{id}"
    combined_question = ' '.join(dataframe[dataframe['id'] == id_value]['질문'])
    rows_to_drop = dataframe[dataframe['id'] == id_value].index[1:]
    dataframe = dataframe.drop(rows_to_drop)
    dataframe.loc[dataframe['id'] == id_value, '질문'] = combined_question
    return dataframe

new_test = update_questions(new_test, '128')
new_test
```
