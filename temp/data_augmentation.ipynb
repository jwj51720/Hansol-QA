# papago augmentation

```python
import requests
import json
 
CLIENT_ID, CLIENT_SECRET = '4TR6H49rGsR0xKUYIs4T', '3W3zj0J_hu'
```

```python
class Papago:
    def __init__(self, CLIENT_ID, CLIENT_SECRET):
        self.url = 'https://openapi.naver.com/v1/papago/n2mt'
        self.headers = {
            'Content-Type': 'application/json',
            'X-Naver-Client-Id': CLIENT_ID,
            'X-Naver-Client-Secret': CLIENT_SECRET
        }

    def translate(self, source, target, text):
        data = {
            'source': source,
            'target': target,
            'text': text
        }
        response = requests.post(self.url, json=data, headers=self.headers)
        return response.json()['message']['result']['translatedText']
    def back_translate(self, path, text):
        for i in range(len(path) - 1):
            data = {'source': path[i], 'target': path[i + 1], 'text': text}
            response = requests.post(self.url, json=data, headers=self.headers)
            if 'errorMessage' in response.json():
                print(response.json())
            text = response.json()['message']['result']['translatedText']
            # print(i, ":", text)
        return text
```
